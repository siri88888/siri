```python
from pypdf import PdfReader

reader = PdfReader("sample-invo.pdf")
page = reader.pages[0]
string=page.extract_text()
```

```python
pattern="[a-zA-Z]+\:\s\+\d+\s\d+\s\d+\-\d"
```

```python
import re
```

```python
var=re.findall(pattern,string)
```

```python
var[0]
```

```python
print(re.findall(pattern1,string))
```
