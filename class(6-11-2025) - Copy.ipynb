```python
def valid_email(email):
    pattern =  r'^[a-z][a-zA-Z0-9._]+@[a-zA-Z0-9]+\.[a-zA-Z]+$'
    if re.match(pattern, email):
        print(" Valid email address")
    else:
        print("Invalid email address")
    
email ="mayur.kulakrni@hotmail.com"
valid_email(email)
```
