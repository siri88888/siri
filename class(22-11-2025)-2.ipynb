```python
def min_max_recursive(lst,low,high):
    if low==high:
        return lst[low],lst[low]
    elif high==low+1:
        if lst[low]<lst[high]:
            min_ele=lst[low]
            max_ele=lst[high]
        else:
            min_ele=lst[high]
            max_ele=lst[low]
        return min_ele,max_ele
    mid=(low+high)//2
    left_min,left_max=min_max_recursive(lst,low,mid)
    right_min,right_max=min_max_recursive(lst,mid+1,high)
    original_min=smaller(left_min,right_min)
    original_max=bigger(left_max,right_max)
    return original_min,original_max
def smaller(a,b):
    if a<b:
        return a
    else:
        return b
def bigger(a,b):
    if a>b:
        return a
    else:
        return b
lst=[23,11,-9,7,34,16,-18,45,33,27,-20]
low=0
high=len(lst)-1
min_max_recursive(lst,low,high)
    
```

```python
def binray_search_recursion(lst,ele):
    mid=len(lst)//2
    if len(lst)<=0:
        return False
    elif lst[mid]==ele:
        return True
    elif lst[mid]>ele:
        return binray_search_recursion(lst[0:mid],ele)
    else:
        return binray_search_recursion(lst[mid+1:],ele)
lst=[2,11,13,19,21,27,45,89,90,100]
key=27
binray_search_recursion(lst,key)
```

```python
def Binary_search(lst,low,high,key):
    if low>high:
        return -1
    mid=(low+high)//2
    if lst[mid]==key:
        return mid
    elif lst[mid]>key:
        return Binary_search(lst,low,mid,key)
    else:
        return Binary_search(lst,mid+1,high,key)
lst=[2,11,13,19,21,27,45,89,90,100]
key=27
Binary_search(lst,key)
```

```python
def inversion(lst,low,high,count=0):
    if low==high:
        return count
    mid=(low+high)//2
    if lst[mid]>lst[mid+1]:
        count+=1
    left=inversion(lst,low,mid,count)
    right=inversion(lst,mid+1,high,count)
    return left+right-count

lst = [13,7,9,5,11,23,16]
inversion(lst, 0, len(lst) - 1)
```
