```python
from abc import ABC,abstractmethod
class shape(ABC):
    def __init__(self):
        pass
    @abstractmethod
    def area(self):
        pass
class circle(shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        print("area is:",3.142*self.radius*self.radius)
class rectangle(shape):
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def area(self):
        print("area is:",self.length*self.breadth)
    
    
class traiangle(shape):
    def __init__(self,base,height):
        self.base=base
        self.height=height
    def area(self):
        print("area is:",0.5*self.base*self.height)
    
    
```

```python
class Shape:
    def __init__(self):
        pass
    def area(self):
        pass
class Circle(Shape):
    def __init__(self):
        pass
    def area(self,radius):
        self.radius=radius
        print("circle radius = ",3.14*self.radius*self.radius)
class Rectange(Shape):
    def __init__(self):
        pass
    def area(self,l,b):
        self.l=l
        self.b=b
        print("ractange area=",self.l*self.b)
c = Circle()
c.area(3)
s = Rectange()
s.area(3,4)
```
