```python
def mode(lst):
    count1=0
    count2=0
    for i in lst:
        if i=="happy":
            count1+=1
        else:
            count2+=1
        print("happy ",count1)
        print("unhappy ",count2)
```

```python
lst=["happy", "happy","unhappy","happy","unhappy"] #find out its mode
def mode(lst):
    count1=0
    count2=0
    for i in lst:
        if i=="happy":
            count1+=1
        else:
            count2+=1
    if count1>count2:
        print("happy ",count1)
    else:
        print("unhappy ",count2)
mode(lst)
```
