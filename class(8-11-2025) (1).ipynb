```python
f=open("lp.txt","r")
alpha=0
digits=0
sc=0
string=f.read()
for ch in string:
    if ch.isalpha():
        alpha+=1
    elif ch.isdigit():
        digits+=1
    else:
        sc+=1
print("alphabets:",alpha)
print("digit:",digits)
print("special_characters:",sc)
f.close()
```

```python
f=open("lp.txt","r")
data =f.read()

newdata=""
for x in data:
    newdata=x
    if ((x>='a' and x<='z') or (x>='A' and x<='Z')):
        newdata=newdata.replace(x,'#')
    elif x>='0' and x<='9':
        newdata=newdata.replace(x,'*')
    else:
        newdata=newdata.replace(x,'%')
    print(newdata,end="")
    if(x=='\n'):
        print("\n")
```
