```python
class circle:
    def __init__(self,radius):
        self.radius=radius
    def calculate(self):
        print(3.142*self.radius*self.radius)
```

```python
obj=circle(14)
```

```python
obj.calculate()
```

```python
class Rectangle:
    def __init__(self,length,breadth):
        self.length=length
        self.breadth=breadth
    def area(self):
        print(self.length*self.breadth)
```

```python
obj=Rectangle(14,15)
```

```python
obj.area()
```

```python
class stackoperations:
    def __init__(self):
        self.stack=[]
    def push(self,element):
        self.stack.append(element)
        return self.stack
    def pop(self):
        self.stack.pop(-1)
        return self.stack
        
```

```python
s=stackoperations()
```

```python
s.push(10)
```

```python
s.push(20)
```
