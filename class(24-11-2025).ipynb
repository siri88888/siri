```python
def matrix_Addition(m1,m2):
    c=[[0,0],[0,0]]
    for i in range(0,len(m1)):
        for j in range(0,len(m1[0])):
            c[i][j]=m1[i][j]+m2[i][j]
    return c
m1=[[1,2],[3,4]]
m2=[[1,2],[3,4]]
matrix_Addition(m1,m2)
            
```
