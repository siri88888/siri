```python
class academics:
    def __init__(self, grade, subjects):
        self.grade=grade
        self.subjects=subjects

class extraCircular:
    def __init__(self, activities, awards):
        self.activities=activities
        self.awards=awards

class student(academics, extraCircular):
    def __init__(self, id, age, name,grade, subjects,activities, awards):
        self.id=id
        self.name=name
        self.age=age
        academics.__init__(self, grade, subjects)
        extraCircular.__init__(self, activities, awards)
class underGraduateStudent(student):
    def __init__(self, subjects, marks, name, age):
        super().__init__(name, age)
        self.subjects=subjects
        self.marks=marks

    def display(self):
        print(self.name)
        print(self.age)
        print(self.subjects)
        print(self.marks)

class graduateStudent(student):
    def __init__(self, subjects1, marks1, name, age):
        super().__init__(name, age)
        self.subjects1=subjects1
        self.marks1=marks1

    def display(self):
        print(self.name)
        print(self.age)
        print(self.subjects1)
        print(self.marks1)
```
